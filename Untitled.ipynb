```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 3, 50)
print(x, len(x))
y = np.exp(x)
y2 = np.exp(2*x)

plt.plot(x, y, label="y = exp(x)")
plt.plot(x, y2, label="y = exp(2x)")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Exponential curve')

plt.legend()
plt.grid()

plt.show()
```
